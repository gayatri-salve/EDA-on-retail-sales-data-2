```python
# Customer and Product Analysis

# 1. Nutritional Content
plt.figure(figsize=(10, 6))
sns.barplot(x='Item', y='Calories', data=data)
plt.title('Calories per Item')
plt.show()

# 2. Product Variety
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=data)
plt.title('Product Variety')
plt.show()

# 3. Target Audience
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Calories', y='Sodium', data=data)
plt.title('Calories vs Sodium')
plt.show()

# Visualization

# 1. Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Item', y='Total Fat', data=data)
plt.title('Total Fat per Item')
plt.show()

# 2. Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Serving Size', y='Calories', data=data)
plt.title('Calories vs Serving Size')
plt.show()

# 3. Heatmap
# Select specific columns
corr_cols = ['Calories', 'Total Fat', 'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']

# Calculate the correlation
corr = data[corr_cols].corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
